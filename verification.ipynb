```python
# plot wave elevation of a sloshing wave: ALE-FEM (quad) vs ALE-FEM (tri) vs EL-BEM
import numpy as np
from matplotlib import pyplot as plt
wav_quad = np.loadtxt("wav_quad.dat", comments="#", delimiter=" ", unpack=False)
wav_tri = np.loadtxt("wav_tri.dat", comments="#", delimiter=" ", unpack=False)
wav_bem = np.loadtxt("wav_bem_10.dat", comments="#", delimiter=" ", unpack=False)

plt.figure(1)
plt.figure(figsize=(12,6))
plt.plot(wav_quad[:,0],wav_quad[:,1],'-r',wav_tri[:,0],wav_tri[:,1],'-b',wav_bem[:,0],wav_bem[:,1],'--k')
plt.legend(['ALE-FEM-quadrilateral element','ALE-FEM-quadrilateral element','ALE-FEM-triangular element','EL-BEM'])
plt.xlabel('time (s)')
plt.ylabel('wave elevatoin (m)')
```
